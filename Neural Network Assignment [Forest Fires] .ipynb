```python
Forest_Fires = pd.read_csv('Forest Fires.csv')
```

```python
model_data=Forest_Fires[['DMC','DC','RH','rain','wind','area','monthapr', 'monthaug', 'monthdec', 
               'monthfeb', 'monthjan', 'monthjul', 'monthjun', 'monthmar', 'monthmay', 
               'monthnov', 'monthoct', 'monthsep','size_category']]
model_data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(model_data["size_category"])
model_data["size_category"]=le.transform(model_data["size_category"])
```

```python
X = model_data.drop('size_category',axis=1)
y = model_data[['size_category']]
```

#### Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3, stratify=y)
X_train.shape,X_test.shape, y_train.shape, y_test.shape
```

#### Build the Model

```python
from keras.optimizers.legacy import SGD
import sklearn
```

```python
def create_model():
    model1 = Sequential()
    model1.add(Dense(64,input_dim=18,activation='relu'))
    model1.add(Dense(64,activation='relu'))
    model1.add(Dense(32,activation="relu"))
    model1.add(Dense(units=1,activation="sigmoid"))
    
    adam = Adam(learning_rate=0.01)
    model1.compile(loss='binary_crossentropy',optimizer = adam,metrics=["accuracy"])
    return model1
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV, KFold
```

```python
model1 = KerasClassifier(build_fn= create_model, verbose=0)
batch_size = [10,20,40,50]
epochs = [10,50,100,200]
param_grid = dict(batch_size= batch_size, epochs=epochs)
grid = GridSearchCV(estimator= model1, param_grid= param_grid, cv= KFold(), verbose= 10)
```

```python
grid_result = grid.fit(X_test,y_test)
```

```python
print('Best {}, using {}'.format(grid_result.best_score_,grid_result.best_params_))
means = grid_result.cv_results_["mean_test_score"]
stds = grid_result.cv_results_["std_test_score"]
params = grid_result.cv_results_["params"]
for mean,stdev,param in zip(means,stds,params):
    print("{}, {} with {}".format(mean,stdev,param))
```

```python
model = Sequential()
model.add(Dense(units=50, input_dim=X_train.shape[1], kernel_initializer='normal', activation='tanh'))
model.add(Dense(units=6, kernel_initializer='normal', activation='tanh'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

model.compile(optimizer='Adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
FF_Model= model.fit(X_train,y_train, validation_data=(X_test,y_test), epochs=50, batch_size=10 )
```

```python
scores = model.evaluate(X,y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
FF_Model.history.keys()
```

#### Compare the Train-Set and Test-Set Accuracy

```python
# Summarize History for Accuracy
plt.figure(figsize=(9,6))
plt.plot(FF_Model.history['accuracy'])
plt.plot(FF_Model.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Summarize History for loss
plt.figure(figsize=(9,6))
plt.plot(FF_Model.history['loss'])
plt.plot(FF_Model.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

#### Predicting Values and Calculate Metrics

```python
from sklearn.metrics import confusion_matrix, classification_report, auc, roc_auc_score, roc_curve
```

```python
# generating predictions for test data
y_predict_test = model.predict(X_test) 

# creating table with test price & predicted price for test
test_prediction = pd.DataFrame()
test_prediction['Test_Actual'] = y_test.size_category
test_prediction['Test_Probability'] = y_predict_test
def probToBinary(varProb):
    if varProb >= 0.5:
        return 1
    else:
        return 0

# converting the probability of target variable to binary class of test data
test_prediction['Test_Predicted'] = test_prediction['Test_Probability'].apply(probToBinary)
print(test_prediction.shape)
test_prediction.head(10)
```

```python
# generating predictions for train data
y_predict_train = model.predict(X_train)

# creating table with test price & predicted price for test
train_prediction = pd.DataFrame()
train_prediction['Train_Actual'] = y_train.size_category
train_prediction['Train_Probability'] = y_predict_train
train_prediction['Train_Predicted'] = train_prediction['Train_Probability'].apply(probToBinary)
print(train_prediction.shape)
train_prediction.head(10)
```

```python
cm = confusion_matrix(train_prediction['Train_Actual'],train_prediction['Train_Predicted'])
sns.heatmap(cm, annot=True)
plt.show()
```

```python
print(classification_report(train_prediction['Train_Actual'],train_prediction['Train_Predicted']))
```

```python
cm = confusion_matrix(test_prediction['Test_Actual'], test_prediction['Test_Predicted'])
sns.heatmap(cm, annot=True)
plt.show()
```

```python
print(classification_report(test_prediction['Test_Actual'], test_prediction['Test_Predicted']))
```
